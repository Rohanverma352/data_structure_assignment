1. Discuss string slicing in python and provide examples.
String slicing in Python is a powerful feature that allows you to extract a subset of characters from a string. Here's a detailed explanation with examples:

Basic Syntax

The syntax for string slicing is:

string[start:stop:step]

Where:

- start: The starting index (inclusive)
- stop: The ending index (exclusive)
- step: The increment between indices (default is 1)

```python
#Examples

#1. Basic Slicing

my_string = "Hello, World!"
print(my_string[0:5])  # Output: "Hello"

#Extracts the first 5 characters.

#1. Omitting Start Index

my_string = "Hello, World!"
print(my_string[:5])  # Output: "Hello"

#Starts from the beginning of the string.

#1. Omitting Stop Index

my_string = "Hello, World!"
print(my_string[7:])  # Output: "World!"

#Goes until the end of the string.

#1. Negative Indices

my_string = "Hello, World!"
print(my_string[-5:])  # Output: "World"

#Starts from the 5th character from the end.

#1. Step Parameter

my_string = "Hello, World!"
print(my_string[::2])  # Output: "Hlo ol!"

#Extracts every 2nd character.

#1. Reversing a String

my_string = "Hello, World!"
print(my_string[::-1])  # Output: "!dlroW ,olleH"

#Reverses the entire string.

#1. Slicing with Negative Step

my_string = "Hello, World!"
print(my_string[5::-1])  # Output: "olleH"

#Extracts characters from index 5 to the beginning, in reverse order.

#Common Use Cases

#1. Extracting substrings
#2. Validating input strings
#3. Parsing data from strings
#4. Creating string templates
```

2. Explain the key features of lists in python.
Lists are a fundamental data structure in Python, offering powerful features for storing and manipulating collections of data. Here are the key features of lists in Python:

Key Features:

1. Ordered Collection: Lists maintain the order of elements.

2. Indexed: Elements are accessible via zero-based indexing.

3. Mutable: Lists can be modified after creation.

4. Dynamic Size: Lists can grow or shrink dynamically.

List Operations:

1. Indexing: Accessing elements using square brackets [].

Example: my_list = [1, 2, 3]; print(my_list[0]) # Output: 1

2. Slicing: Extracting subsets using start:stop:step.

Example: my_list = [1, 2, 3, 4, 5]; print(my_list[1:3]) # Output: [2, 3]

3. Append: Adding elements using append() or extend().

Example: my_list = [1, 2]; my_list.append(3); print(my_list) # Output: [1, 2, 3]

4. Insert: Inserting elements at specific positions using insert().

Example: my_list = [1, 2]; my_list.insert(1, 3); print(my_list) # Output: [1, 3, 2]

5. Remove: Removing elements using remove() or pop().

Example: my_list = [1, 2, 3]; my_list.remove(2); print(my_list) # Output: [1, 3]

6. Sort: Sorting elements using sort() or sorted().

Example: my_list = [3, 2, 1]; my_list.sort(); print(my_list) # Output: [1, 2, 3]

7. Iterate: Iterating over elements using for loops.

Example: my_list = [1, 2, 3]; for elem in my_list: print(elem) # Output: 1, 2, 3


List Methods:

1. append()
2. extend()
3. insert()
4. remove()
5. pop()
6. index()
7. count()
8. sort()
9. reverse()
10. clear()


List Comprehensions:

A concise way to create lists.

Example: squares = [x**2 for x in range(5)]; print(squares) # Output: [0, 1, 4, 9, 16]

```python
#3. Describe how to access, modify, and Delete elements in a list with examples.

#explanation of accessing, modifying, and deleting elements in a list:

#Accessing Elements:

#1. Indexing: Use square brackets [] with the index.

my_list = [1, 2, 3, 4, 5]
print(my_list[0])  # Output: 1

#2. Negative Indexing: Use -1 to access the last element.

my_list = [1, 2, 3, 4, 5]
print(my_list[-1])  # Output: 5

#3. Slicing: Use start:stop:step to access a subset.

my_list = [1, 2, 3, 4, 5]
print(my_list[1:3])  # Output: [2, 3]

#Modifying Elements:

#. Assignment: Assign a new value to an index.

my_list = [1, 2, 3, 4, 5]
my_list[0] = 10
print(my_list)  # Output: [10, 2, 3, 4, 5]

#2. Append: Add an element to the end.

my_list = [1, 2, 3]
my_list.append(4)
print(my_list)  # Output: [1, 2, 3, 4]

#3. Insert: Insert an element at a specific position.

my_list = [1, 2, 3]
my_list.insert(1, 4)
print(my_list)  # Output: [1, 4, 2, 3]

#4. Extend: Add multiple elements.

my_list = [1, 2, 3]
my_list.extend([4, 5, 6])
print(my_list)  # Output: [1, 2, 3, 4, 5, 6]

#Deleting Elements:

#1. Remove: Remove the first occurrence.

my_list = [1, 2, 3, 4, 2]
my_list.remove(2)
print(my_list)  # Output: [1, 3, 4, 2]

#2. Pop: Remove and return an element.

my_list = [1, 2, 3]
print(my_list.pop(0))  # Output: 1
print(my_list)  # Output: [2, 3]

#3. Del: Delete an element or slice.

my_list = [1, 2, 3]
del my_list[0]
print(my_list)  # Output: [2, 3]

#4. Clear: Remove all elements.

my_list = [1, 2, 3]
my_list.clear()
print(my_list)  # Output: []

#Additional Methods:

#1. Index: Find the index of an element.

my_list = [1, 2, 3]
print(my_list.index(2))  # Output: 1

#2. Count: Count occurrences of an element.

my_list = [1, 2, 2, 3]
print(my_list.count(2))  # Output: 2

#3. Sort: Sort the list.

my_list = [3, 2, 1]
my_list.sort()
print(my_list)  # Output: [1, 2, 3]

#4. Reverse: Reverse the list.

my_list = [1, 2, 3]
my_list.reverse()
print(my_list)  # Output: [3, 2, 1]
```

```python
#5. Describe the key features of sets and provide examples of their use.

#Sets are a fundamental data structure in Python, providing an efficient way to store and manipulate unique elements.

#Key Features:

#1. Unordered collection
#2. Unique elements (no duplicates)
#3. Mutable (can be modified)
#4. Fast membership testing (in, not in)
#5. Support for mathematical operations (union, intersection, difference)

#Creating Sets:

#1. Using the set() function:

my_set = set([1, 2, 3, 2, 1])  # my_set = {1, 2, 3}

#2. Using curly brackets {}:

my_set = {1, 2, 3, 2, 1}  # my_set = {1, 2, 3}

#Set Operations:

#1. Union (| or union()):

set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(set1 | set2)  # Output: {1, 2, 3, 4, 5}

#2. Intersection (& or intersection()):

set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(set1 & set2)  # Output: {3}

#3. Difference (- or difference()):

set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(set1 - set2)  # Output: {1, 2}

#4. Symmetric Difference (^ or symmetric_difference()):

set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(set1 ^ set2)  # Output: {1, 2, 4, 5}

#Set Methods:

#1. add(): Add an element.

my_set = {1, 2}
my_set.add(3)
print(my_set)  # Output: {1, 2, 3}

#2. remove(): Remove an element.

my_set = {1, 2, 3}
my_set.remove(2)
print(my_set)  # Output: {1, 3}

#3. discard(): Remove an element if present.

my_set = {1, 2, 3}
my_set.discard(4)  # No error

#4. clear(): Clear the set.

my_set = {1, 2, 3}
my_set.clear()
print(my_set)  # Output: set()

#5. copy(): Create a copy.

my_set = {1, 2, 3}
new_set = my_set.copy()
print(new_set)  # Output: {1, 2, 3}

#Use Cases:

#1. Removing duplicates from a list.
#2. Fast membership testing.
#3. Mathematical operations (union, intersection, difference).
#4. Data deduplication.
#5. Set-based data structures (e.g., graphs, networks).
```

```python
#7. Describe how to add, modify, and delete items in a dictionary with examples

#how to add, modify, and delete items in a dictionary:

#Adding Items:

#1. Direct Assignment:

my_dict = {"name": "John", "age": 30}
my_dict["city"] = "New York"
print(my_dict)  # Output: {"name": "John", "age": 30, "city": "New York"}

#2. update() Method:

my_dict = {"name": "John", "age": 30}
my_dict.update({"city": "New York", "country": "USA"})
print(my_dict)  # Output: {"name": "John", "age": 30, "city": "New York", "country": "USA"}

#3. setdefault() Method:

my_dict = {"name": "John", "age": 30}
my_dict.setdefault("city", "New York")
print(my_dict)  # Output: {"name": "John", "age": 30, "city": "New York"}

#Modifying Items:

#1. Direct Assignment:

my_dict = {"name": "John", "age": 30}
my_dict["age"] = 31
print(my_dict)  # Output: {"name": "John", "age": 31}

#2. update() Method:

my_dict = {"name": "John", "age": 30}
my_dict.update({"age": 31})
print(my_dict)  # Output: {"name": "John", "age": 31}

#Deleting Items:

#1. del Statement:

my_dict = {"name": "John", "age": 30}
del my_dict["age"]
print(my_dict)  # Output: {"name": "John"}

#2. pop() Method:

my_dict = {"name": "John", "age": 30}
my_dict.pop("age")
print(my_dict)  # Output: {"name": "John"}

#3. popitem() Method (removes last inserted item):

my_dict = {"name": "John", "age": 30}
my_dict.popitem()
print(my_dict)  # Output: {"name": "John"}

#4. clear() Method (removes all items):

my_dict = {"name": "John", "age": 30}
my_dict.clear()
print(my_dict)  # Output: {}

#Additional Methods:

#1. get() Method (retrieves value):
    
my_dict = {"name": "John", "age": 30}
print(my_dict.get("name"))  # Output: John

#2. items() Method (returns key-value pairs):

my_dict = {"name": "John", "age": 30}
print(my_dict.items())  # Output: dict_items([('name', 'John'), ('age', 30)])

#3. keys() Method (returns keys):

my_dict = {"name": "John", "age": 30}
print(my_dict.keys())  # Output: dict_keys(['name', 'age'])

#4. values() Method (returns values):

my_dict = {"name": "John", "age": 30}
print(my_dict.values())  # Output: dict_values(['John', 30])
```

```python
#8. Discuss the importance of dictionary keys being immutable and provide examples

#dictionary keys must be immutable, meaning they cannot be changed after creation. This requirement ensures efficient and reliable dictionary operations.


#Why Immutability Matters:

#1. Hashing: Dictionaries rely on hashing to store and retrieve key-value pairs efficiently. Immutable keys ensure consistent hash values, enabling fast lookups.
#2. Uniqueness: Immutable keys prevent accidental modifications, ensuring each key remains unique.
#3. Thread Safety: Immutable keys make dictionaries thread-safe, as multiple threads can access the dictionary without modifying keys.

#Examples of Immutable Keys:

#1. Strings:

my_dict = {"name": "John", "age": 30}

#2.  Integers:

    #python
my_dict = {1: "John", 2: "Jane"}

#3. Tuples (if elements are immutable):

my_dict = {(1, 2): "John", (3, 4): "Jane"}

#4.  Frozensets:

    #python
my_dict = {frozenset([1, 2]): "John", frozenset([3, 4]): "Jane"}

#Examples of Mutable Keys (Invalid):

#5. Lists:

#Invalid: Lists are mutable

my_dict = {[1, 2]: "John", [3, 4]: "Jane"}

#6.  Dictionaries:
   
# Invalid: Dictionaries are mutable

my_dict = {{1: 2}: "John", {3: 4}: "Jane"}

#7. Sets:

#Invalid: Sets are mutable

my_dict = [{1, 2}: "John", {3, 4}: "Jane"]

#Consequences of Using Mutable Keys:#

#1. TypeError: Attempting to use mutable keys raises a TypeError.
#2. Inconsistent Behavior: Modifying keys can lead to unexpected dictionary behavior.
#3. Performance Issues: Mutable keys can degrade dictionary performance.
```
